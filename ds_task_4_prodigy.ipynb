```python
usa = df[df.Country == " USA"]
```

```python
uk = df[df.Country == " UK "]
```

```python
usa.head()
```

```python
uk.head()
```

```python
plt.plot(usa.Day,usa.Hour,'o')
plt.plot(uk.Day,uk.Hour,'*')
plt.xlabel("Day")
plt.ylabel("Hour")
plt.title("Day vs Hour")
plt.show()
```

```python
plt.plot(usa.Year,usa.Month,'o')
plt.plot(uk.Year,uk.Month,'*')
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Year vs Month")
plt.show()
```

```python
plt.bar(usa.Retweets,usa.Likes)
plt.bar(uk.Retweets,uk.Likes)
plt.xlabel("Retweets")
plt.ylabel("Likes")
plt.title("Retweets vs Likes")
plt.show()
```

```python
plt.hist(uk.Platform)
plt.hist(usa.Platform)
plt.title("Platform")
plt.show()
```

```python
plt.scatter(usa.Year,usa.Year)
plt.scatter(uk.Month,uk.Day)
plt.title("Month vs Year")
plt.show()
```
